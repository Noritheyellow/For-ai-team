```python
import os
import gzip
import wfdb
import pickle
import random
import numpy as np
import pandas as pd

DS_PATH = '/root/Workspace/DataWarehouse/mimic-iii_wfdb/PPG-ABP'
```

```python
SUBDIR = []
subdir_nrec = os.scandir(f'{DS_PATH}/01_subdir_records')
for rec in subdir_nrec:
    with gzip.open(f'{rec.path}') as f:
        SUBDIR.append(np.asarray(pickle.load(f)))

for sub in SUBDIR:
    print(f'Total {len(SUBDIR)} sub-directories exist. For instance, first sub-dir has {sub.shape[0]} records loaded.')
```

```python
# get 1 sample
samples = random.sample(list(SUBDIR[0]), k=1)
samp_sig, samp_fields = wfdb.rdsamp(samples[0][:-4])
samp_num = wfdb.rdsamp('../../DataLake/mimic-iii_wfdb/p00/p000020/p000020-2183-04-28-17-47n')
pd.DataFrame(samp_num[0], columns=samp_num[1]['sig_name']).head(25)
```
